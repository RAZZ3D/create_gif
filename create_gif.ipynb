```python
import os
import imageio.v3 as iio
```

```python
filenames = ['evo1.jpg', 'evo2.jpg']
directory = r'C:\Users\RAZZ3D'
images = [ ]
```

```python
for filename in filenames:
    full_path = os.path.join(directory, filename)  # safely builds the full path
    images.append(iio.imread(full_path))
```

```python
output_path = os.path.join(directory,'EVO.gif')
iio.imwrite(output_path, images, duration=500, loop=0)
```
